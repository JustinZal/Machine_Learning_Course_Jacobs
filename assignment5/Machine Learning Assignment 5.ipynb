```python
import numpy as np
import matplotlib.pyplot as plt


#Task 1

y = np.array([0, 0.3, 0.75, 1, 2])
X = np.array([
    [1, 0, 0],
    [1, 1, 2],
    [1, 2, 4],
    [1, 3, 0],
    [1, 4, 1]
])

term1 = (X.T).dot(y)
term2 = np.linalg.inv((X.T).dot(X))
prediction_coefficients = term2.dot(term1)

def predict(coeff, x, y):
    return coeff[0] + coeff[1] * x + coeff[2] * y

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = predict(prediction_coefficients, X, Y)

fig = plt.figure()
ax = fig.gca(projection='3d')

temp = ax.plot_surface(X, Y, Z)
```

```python
#Task 2 Energy efficiency
import numpy as np
import pandas

raw_dataset = pandas.read_excel('./ENB2012_data.xlsx', usecols=[f'X{i}' for i in range(1, 9)] + ['Y1'])

inputs = []
outputs = []
for _, row in raw_dataset.iterrows():
    row_list = row.tolist()
    inputs.append([1] + row_list[0:-1])
    outputs.append(row[-1])

  
X = np.array(inputs)
y = np.array(outputs)

term1 = (X.T).dot(y)
term2 = np.linalg.inv((X.T).dot(X))
coefficients = term2.dot(term1)

coefficients = np.linalg.inv((X.T).dot(X)).dot((X.T).dot(y))

#Making predictions for 1st 3 rows

def predict_heating(input_array):
    return np.append([1], input_array).dot(coefficients)

print('Prediction 1:', predict_heating(np.array([0.98, 514.50, 294.00, 110.25, 7.00, 2, 0.00, 0])))
print('Prediction 2:', predict_heating(np.array([0.98, 514.50, 294.00, 110.25, 7.00, 3, 0.00, 0])))
print('Prediction 3:', predict_heating(np.array([0.98, 514.50, 294.00, 110.25, 7.00, 4, 0.00, 0])))
```
