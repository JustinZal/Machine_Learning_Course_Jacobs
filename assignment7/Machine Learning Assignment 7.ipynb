```python
import numpy as np
import matplotlib.pyplot as plt

N = 51

input_data = np.random.uniform(-4, 4, N)
output_data = input_data ** 2 + np.random.normal(0, 1.15, N)

data = np.array([[input_data[i], output_data[i]] for i in range(N)])


def l2_norm(x1, x2):
    return (x1 - x2) ** 2


def compute_error(pred_input, prediction_set, result, k):
    closest = sorted(prediction_set, key=lambda point: l2_norm(point[0], pred_input))[0:k]
    prediction = np.array([point[1] for point in closest]).mean()
    
    return l2_norm(prediction, result)


def K_fold_cross_validation(k, dataset, K):  
    splits = np.array_split(dataset, K)
    
    errors = []
    for i in range(len(splits)):
        val = splits[i].tolist()
        test = np.concatenate(splits[0:i] + splits[i+1:])
        err = 0
        
        for j in range(len(val)):
            err += compute_error(val[j][0], test, val[j][1], k)
            
        errors.append(err)
    
    return np.array(errors).mean()


def validation_set_error(dataset, k):
    np.random.shuffle(dataset)
    train = dataset[0:41]
    val = dataset[41:]
    
    return sum([compute_error(point[0], train, point[1], k) for point in val]) / len(val)

k_set = [k for k in range(2, 21)]

K_fold_errors = [K_fold_cross_validation(k, data, 10) for k in k_set]
leave_one_cross_out_errors = [K_fold_cross_validation(k, data, N) for k in k_set]
validation_set_errors = [validation_set_error(data, k) for k in k_set]

plt.plot(k_set, K_fold_errors, k_set, leave_one_cross_out_errors, k_set, validation_set_errors)
plt.legend(['K fold validation', 'leave one cross out validation', 'validation set method'])
plt.xlabel('k')
plt.ylabel('Generalization Error')
plt.show()
```
