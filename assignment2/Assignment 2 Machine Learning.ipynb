```python
mu = np.array([0, 1])
sigma = np.matrix([[1, 0.25], [0.5, 1]])

def gaussian_distribution(x):
    k = 1 / np.sqrt((2 * np.pi) ** 2 * np.linalg.det(sigma))
    
    inverse = np.linalg.inv(sigma)
    matrix_product = np.matmul(inverse, x - mu)
    dot_product = (x - mu).dot(np.asarray(matrix_product).reshape(-1))
    
    return k.item() * np.exp(-0.5 * dot_product).item()
    
x = np.linspace(-12, 12, 300)
y = np.linspace(-12, 12, 300)

z = []
for i in range(300):
    for j in range(300):
        z.append(gaussian_distribution(np.array([x[i], y[j]])))

x , y = np.meshgrid(x,y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
z = np.array(z).reshape(300, 300)

ax.plot_surface(x,y,z)
plt.show()
```
